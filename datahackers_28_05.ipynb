```python
# contador de pontos para homens e mulheres

countMale = 0
countFemale = 0
countTied = 0

def countPoints(valueMale, valueFemale):
  global countMale
  global countFemale
  global countTied

  if (valueMale > valueFemale):
    countMale += 1

  elif (valueMale < valueFemale):
    countFemale += 1

  else: countTied += 1
  
  print(f'H: {countMale}, M: {countFemale}, Tied: {countTied}')
```

```python
manLiveBrazil = banco['cliente_ID'].loc[(banco['live_Brazil']== 1) & (banco['Sexo']=='Masculino')].count()
print(f'Homens que residem no Brasil : {round(manLiveBrazil, 2)}')

womanLiveBrazil = banco['cliente_ID'].loc[(banco['live_Brazil']== 1) & (banco['Sexo']=='Feminino')].count()
print(f'Mulheres que residem no Brasil : {round(womanLiveBrazil, 2)}')

banco['cliente_ID'].loc[banco['live_Brazil'] == 1].groupby (banco['Sexo']).count().plot(kind = 'bar')

countPoints(manLiveBrazil, womanLiveBrazil)
```
